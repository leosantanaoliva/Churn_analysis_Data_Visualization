```python
import pandas as pd

# Load the dataset
file_path = '/Users/leonardosantana/Directory/Project3/telco-customer-churn/WA_Fn-UseC_-Telco-Customer-Churn.csv'
df = pd.read_csv(file_path)

# Display the first few rows of the dataset
df.head()
```

```python
# Check for missing values
missing_values = df.isnull().sum()

df['MultipleLines'] = df['MultipleLines'].astype(int)

# Convert Remaining Columns to Numeric if Applicable
numeric_cols = ['TotalCharges','tenure', 'MonthlyCharges']

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Check again for missing values after conversion
missing_values_after_conversion = df.isnull().sum()

# Step 6: Normalize Categorical Values
df['MultipleLines'] = df['MultipleLines'].replace({'No phone service': 'No'})
df['OnlineSecurity'] = df['OnlineSecurity'].replace({'No internet service': 'No'})
df['OnlineBackup'] = df['OnlineBackup'].replace({'No internet service': 'No'})
df['DeviceProtection'] = df['DeviceProtection'].replace({'No internet service': 'No'})
df['TechSupport'] = df['TechSupport'].replace({'No internet service': 'No'})
df['StreamingTV'] = df['StreamingTV'].replace({'No internet service': 'No'})
df['StreamingMovies'] = df['StreamingMovies'].replace({'No internet service': 'No'})

# Fill or remove missing values
df['TotalCharges'].fillna(df['TotalCharges'].mean(), inplace=True)

# Convert SeniorCitizen to categorical type
df['InternetService'] = df['InternetService'].astype('category')
df['Contract'] = df['Contract'].astype('category')
df['PaymentMethod'] = df['PaymentMethod'].astype('category')
df['MultipleLines'] = df['MultipleLines'].astype('category')
df['customerID'] = df['customerID'].astype('category')

# List of service-related columns to consider
service_columns = ['PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 
                   'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 
                   'StreamingMovies']

df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# Count total services subscribed by each customer
df['TotalServices'] = df[service_columns].apply(lambda row: row.value_counts().get('Yes', 0), axis=1)

# Drop duplicates if any
df.drop_duplicates(inplace=True)

# Final clean dataset
clean_df = df

# Display information about the cleaned dataset
clean_df.info()

# Save the Cleaned Dataset
cleaned_file_path = '/Users/leonardosantana/Directory/Project3/telco-customer-churn/cleaned_telco_data.csv'
clean_df.to_csv(cleaned_file_path, index=False)
```

```python
# Columns to replace values
columns_to_replace = ['OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 
                      'StreamingTV', 'StreamingMovies', 'PaperlessBilling', 
                      'Churn', 'PhoneService', 'Dependents', 'MultipleLines','Partner']

# Mapping dictionary
replace_map = {'Yes': 1, 'No': 0}

# Replace values in specified columns
clean_df[columns_to_replace] = clean_df[columns_to_replace].replace(replace_map)

# Display the modified DataFrame
print("DataFrame with replaced values:")
print(clean_df)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report


# Load the dataset
file_path = '/Users/leonardosantana/Directory/Project3/telco-customer-churn/cleaned_telco_data.csv'
df = pd.read_csv(file_path)

df = df.drop('customerID', axis=1, inplace=True)  

# Encode categorical variables
categorical_columns = ['OnlineSecurity', 'OnlineBackup', 'TechSupport', 'Contract', 'Churn']
for column in categorical_columns:
    le = LabelEncoder()
    df[column] = le.fit_transform(df[column])

# Select only the specified columns
selected_columns = ['OnlineSecurity', 'OnlineBackup', 'TechSupport', 'Contract', 'tenure', 'MonthlyCharges', 'Churn']
df_selected = df[selected_columns]

# Compute and visualize the correlation matrix
corr_matrix = df_selected.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```
